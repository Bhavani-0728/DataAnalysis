```python
import pandas as pd

df=pd.read_csv('loan_approved.csv')
df.head(5)
```

Loan_ID -- nominal -- not in a order

Gender	-- nominal -- no order

Married	-- nominal -- no order

Dependents	-- ordinal -- distinct numeric values

Education	-- nominal -- no order

Self_Employed	-- nominal -- no order

ApplicantIncome	-- continuous -- large range of values 

CoapplicantIncome	-- continuous -- large range of values

LoanAmount	-- continuous -- large range of values

Loan_Amount_Term	-- discrete -- multiples of 12

Credit_History	-- discrete

Property_Area	-- ordinal -- in order

Loan_Status (Approved) -- nominal -- no order

```python
df.info()
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df['Gender'].unique()
```

```python
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
df['Married'].unique()
```

```python
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Dependents'].unique()
```

```python
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['Self_Employed'].unique()
```

```python
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].median())
df['LoanAmount'].unique()
```

```python
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
df['Loan_Amount_Term'].unique()
```

```python
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].median())
df['Credit_History'].unique()
```

```python
df.to_csv('loan_approved.csv')
```
