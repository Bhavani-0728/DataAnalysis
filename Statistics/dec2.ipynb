```python
import pandas as pd

df=pd.read_csv('loan_approved.csv')
df.head(5)
```

```python
mean=df['ApplicantIncome'].mean()
print('Mean = ',mean)
std=df['ApplicantIncome'].std()
print('Standard Deviation = ',std)
```

```python
df['ApplicantIncome'].head(5)
```

```python
z1=(5849-mean)/std
print('z1 = ',z1)

z2=(4583-mean)/std
print('z2 = ',z2)

z3=(3000-mean)/std
print('z3 = ',z3)

z4=(2583-mean)/std
print('z4 = ',z4)

z5=(6000-mean)/std
print('z5 = ',z5)
```

```python
max=df['ApplicantIncome'].max()
print('Max = ',max)
min=df['ApplicantIncome'].min()
print('Min = ',min)
range=max-min
print('Range = ',range)
```

```python
df['ApplicantIncome'].value_counts()
```

```python
sk1=df['ApplicantIncome'].skew()
print('Skew = ',sk1)
ku1=df['ApplicantIncome'].kurtosis()
print('Kurtosis = ',ku1)
# highly skewed and leptokurtic --> not acceptable
```

```python
sk2 = df['CoapplicantIncome'].skew()
print(sk2)
ku2 = df['CoapplicantIncome'].kurtosis()
print(ku2)
# highly skewed and leptokurtic --> not acceptable
```
